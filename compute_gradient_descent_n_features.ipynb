```python
def compute_gradient_descent_n_features(X,y,w_in,b_in,alpha,num_itrs,compute_cost,compute_gradient):
    
    J_history = []
    w_history = []
    
    w = w_in
    b = b_in
    
    for i in range(num_itrs):
        dj_dw, dj_db = compute_gradient(X,y,w,b)
        
        w = w - (alpha * dj_dw)
        b = b - (alpha * dj_db)
        
        J_history.append(compute_cost(X,y,w,b))
        w_history.append(w)
        
        if i % (math.ceil(num_itrs)/ 10) == 0:
            print(f"i= {i} Cost = {J_history[-1]}, w = {w_history[-1]}, b = {b_in}")
    
    return w,b,J_history,w_history
```
